```python
def readfiles(filename):

    real_filename = '../books/' + filename
    output_filename = '../data/encoded_books/' + filename
    output_key = '../data/key/' + filename
    data = ''
    # open file wiht binary form
    with open(real_filename, 'r') as file:
        data = file.read()
        return(data)
```

```python
filename = []
for i in range(3001,4000):
    filename.append("%04d.txt"%i)
```

```python
data = list(map(readfiles,filename))
```

```python
import os
 
def listdir(path, list_name):
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        if os.path.isdir(file_path):
            listdir(file_path, list_name)
        elif os.path.splitext(file_path)[1]=='.txt':
            list_name.append(file_path)
```
